```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Embedding,LSTM,MaxPooling1D,Conv1D,Dense,Dropout
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
df=pd.read_csv("Final_cleaned.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df.head()
```

```python
tokenizer=Tokenizer()
tokenizer.fit_on_texts(df['text'])
```

```python
len(tokenizer.word_index)
```

```python
tokenizer=Tokenizer(num_words=20000,oov_token="<oov>")
tokenizer.fit_on_texts(df['text'])
```

```python
tokenizer.word_index
```

```python
X=tokenizer.texts_to_sequences(df['text'])
```

```python
X=pad_sequences(X,maxlen=400)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,df['is_sucidal'].values,test_size=0.3)
```

```python
model=Sequential([

                  Embedding(input_dim=20000,output_dim=128,input_length=400),

                  Conv1D(128,3,padding='same',activation='relu'),
                  MaxPooling1D(),
                  Dropout(0.2),
                  LSTM(256),
                  Dropout(0.2),
                Dense(128,activation='relu',kernel_regularizer=keras.regularizers.L1()),
                  Dense(1,activation='sigmoid')

]
)
```

```python
model.summary()
```

```python
model.compile("adam",loss=tf.losses.binary_crossentropy,metrics=['Accuracy'])
```

```python
hist=model.fit(X_train,y_train,epochs=10,validation_data=(X_test,y_test))
```

```python
model.save("vaibhav.h5")
```

```python
def predict(text):
  text=tokenizer.texts_to_sequences([text])
  text=pad_sequences(text,400)
  temp=model(text)

  if temp>0.5:
    return "Sucidal"
  return "Non Sucidal"
```

```python

predict("hello how are you all today this is a very beautiful day today")
```
