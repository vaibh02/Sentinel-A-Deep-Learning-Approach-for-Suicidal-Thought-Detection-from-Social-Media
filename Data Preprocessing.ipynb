```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)

    return text
```

```python
sucide['text']=sucide['new'].apply(clean_text)
```
